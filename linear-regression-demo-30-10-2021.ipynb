```python
X = df.iloc[:,0:1]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
# prediction
import numpy as np
lr.predict(np.array([[3.5]]))
```

```python
def salary_predictor(years):
    return int(lr.predict(np.array([[years]]))[0])
```

```python
salary_predictor(4)
```

```python
import seaborn as sns
sns.lineplot(df['YearsExperience'],lr.predict(df.iloc[:,0:1]), color='red')
sns.scatterplot(df['YearsExperience'],df['Salary'])
```
